```python
import os
import glob
import re
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
DATA_PATH = os.path.join(os.path.dirname(os.getcwd()), "Data/SuccessfulScrape")
YEAR = 2010
years = {"2010":"2009-10",
         "2011":"2010-11",
         "2012":"2011-12",
         "2013":"2012-13",
         "2014":"2013-14",
         "2015":"2014-15",
         "2016":"2015-16",
         "2017":"2016-17",
         "2018":"2017-18"}
```

```python
def get_single_season_df(year, datapath):
    return pd.read_csv(glob.glob(os.path.join(datapath, f"*_game*_{year}_*"))[0])
```

```python
df = get_single_season_df(YEAR, DATA_PATH)
```

```python
df.head()
```

```python
df.info()
df.describe()
```

```python
df.columns
```

```python
features = list(df.columns[6:-6])
regression = df.columns[-2]
label = df.columns[-1]
cols = features + [regression] + [label]
print(f"features = {features}")
print(f"label = {label}")
print(f"regression = {regression}")
print(f"cols = {cols}")
```

Will attempt to learn who won the game (classification) and how much they won the game by (regression). The game id, team id and abbreviation are unnecessary for the data learning. To trim dataset, I will only use the home team performing learning on both is redundant (if home team wins by 8 then away team loses by 8).

Reorganize/transform the data into a more logical dataframe by using a multi-index for each game and the second index being the home vs away team.

```python
df2 = df.set_index(["game_id", "home_flag"])
df2.head()
```

```python
game_ids = np.array(df2.index.levels[0])
```

Split the data into train test split by splitting the game_ids array.

```python
np.random.seed(42)
np.random.shuffle(game_ids)
```

```python
num_games = len(game_ids)
training = int(0.25*num_games)
validation = int(0.25*num_games) + training
testing = 0.5

training_games = game_ids[:training]
validation_games = game_ids[training:validation]
testing_games = game_ids[validation:]
```

```python
training_df = df2.loc[training_games]
validation_df = df2.loc[validation_games]
testing_df = df2.loc[testing_games]
```

```python
training_subset_home = training_df.query("home_flag == '1'").droplevel(1).loc[:, cols]
training_subset_away = training_df.query("home_flag == '0'").droplevel(1).loc[:, cols]
training_subset = training_subset_home.subtract(training_subset_away)
training_subset["net_score"] = training_subset["net_score"] / 2
training_subset["won"] = (training_subset["won"] + 1)// 2
training_subset.head() 
```

First, let us try and make a classification model to see which team wins, given the input data from the game.

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(solver="newton-cg")

X = training_subset.iloc[:, 1:-2 ].to_numpy()
y = training_subset.iloc[:, -1].to_numpy()

clf.fit(X, y)

clf.score(X, y)
```

```python
validation_subset_home = validation_df.query("home_flag == '1'").droplevel(1).loc[:, cols]
validation_subset_away = validation_df.query("home_flag == '0'").droplevel(1).loc[:, cols]
validation_subset = validation_subset_home.subtract(validation_subset_away)
validation_subset["net_score"] = validation_subset["net_score"] / 2
validation_subset["won"] = (validation_subset["won"] + 1)// 2
validation_subset.head()
```

```python
VX = validation_subset.iloc[:, 1:-2 ].to_numpy()
Vy = validation_subset.iloc[:, -1].to_numpy()

clf.score(VX, Vy)
```

Next I will compare this performance versus picking the team with the better record to see if it produces any better of a result

```python
naiveX = training_subset.loc[:, "wl%"].values
naiveY = np.zeros(len(naiveX))
for i in range(len(naiveX)):
    if naiveX[i] < 0:
        naiveY[i] = 0
    else:
        naiveY[i] = 1
```

```python
y = training_subset.loc[:, "won"].values
correct = 0
for i in range(len(naiveY)):
    if y[i] == naiveY[i]:
        correct += 1

print(correct/len(naiveY))
```

The model appears to perform better than a simplistic decision of picking the team with the better record.

The more interesting test is to see if the model can predict if a team will win or lose based on data going into the game (not data after the game has completed). To test this, we will use the team data for the season and compare the two teams and then predict off this to determine who should win.

```python
testing_df.head()
```

```python
def get_team_df(datapath):
    return pd.read_csv(glob.glob(os.path.join(datapath, "*team*"))[0])
```

```python
team_df = get_team_df(DATA_PATH)
```

```python
team_df.head()
```

```python
team_df_2010 = team_df.loc[team_df["season"] == years[str(YEAR)]]
team_df_2010.set_index("team_id", inplace=True)
team_df_2010 = team_df_2010.loc[:, features]
```

```python
team_df_2010.info()
```

```python
def get_test_input(testing_df, team_df, model):
    X = []
    for game_id in np.unique([index[0] for index in testing_df.index.values]):
        home_team = testing_df.loc[game_id, 1]["team_id"]
        away_team = testing_df.loc[game_id, 0]["team_id"]
        home_team_df = team_df.loc[home_team, features].drop("wl%")
        away_team_df = team_df.loc[away_team, features].drop("wl%")
        data_point = home_team_df.subtract(away_team_df)
        X.append(data_point.to_numpy())
    return X
```

```python
TestX = get_test_input(testing_df, team_df_2010, clf)

Testy = testing_df.query("home_flag == '1'").loc[:, "won"].values

clf.score(TestX, Testy)
```

```python
def test_control_score(testing_df, team_df):
    predicted = []
    for game_id in np.unique([index[0] for index in testing_df.index.values]):
        home_team = testing_df.loc[game_id, 1]["team_id"]
        away_team = testing_df.loc[game_id, 0]["team_id"]
        wl_comp = team_df.loc[home_team, "wl%"] - team_df.loc[home_team, "wl%"]
        if wl_comp < 0:
            predicted.append(0)
        else:
            predicted.append(1)
    Testy = testing_df.query("home_flag == '1'").loc[:, "won"].values
    correct = 0
    for i in range(len(predicted)):
        if Testy[i] == predicted[i]:
            correct += 1
    return correct/len(predicted)
```

```python
print(test_control_score(testing_df, team_df_2010))
```
